```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Step 1: 
np.random.seed(0)
a = np.random.rand(100)
b = np.random.rand(100)
c = np.random.rand(100)
labels = np.array([0.2, 0.5, 1])

data = np.vstack([a, b, c])
print("Original Data:\n", data)

# Step 2: 
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
print("\nScaled Data:\n", scaled_data)

# Step 3-6:  PCA 
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)
print("\nPrincipal Components:\n", principal_components)

# 
components = pca.components_
print("\nPCA Components:\n", components)

# 
top_indices = np.argsort(np.abs(components[0]))[-3:]  # 
print("\nTop contributing wavelengths (indices):\n", top_indices)

# 
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
# sc = ax.scatter(principal_components[:, 0], principal_components[:, 1], principal_components[:, 2], c=labels, cmap='viridis')

# # 
# for i, txt in enumerate(labels):
#     ax.text(principal_components[i, 0], principal_components[i, 1], principal_components[i, 2], f'Label {txt}', fontsize=10)

# ax.set_xlabel('Principal Component 1')
# ax.set_ylabel('Principal Component 2')
# ax.set_zlabel('Principal Component 3')
# ax.set_title('PCA of Spectra Data')

# # 
# cbar = plt.colorbar(sc)
# cbar.set_label('Label')

# plt.show()

# # 
# wavelengths = np.arange(1, 101)
# fig, axs = plt.subplots(3, 1, figsize=(10, 10))
# spectra = [a, b, c]
# colors = ['r', 'g', 'b']
# titles = ['Spectrum A', 'Spectrum B', 'Spectrum C']

# for i in range(3):
#     axs[i].plot(wavelengths, spectra[i], color=colors[i], label=f'Spectrum {chr(65+i)}')
#     for j in top_indices:
#         axs[i].annotate(f'{wavelengths[j]}', (wavelengths[j], spectra[i][j]), textcoords="offset points", xytext=(0,10), ha='center')
#         axs[i].plot(wavelengths[j], spectra[i][j], 'o', color='black')
#     axs[i].set_title(titles[i])
#     axs[i].set_xlabel('Wavelength')
#     axs[i].set_ylabel('Intensity')
#     axs[i].legend()

# plt.tight_layout()
# plt.show()
```
